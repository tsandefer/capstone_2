```python
def get_artist_names(filename):
    # Import data
    with open(filename) as file:
        artist_names = [line.strip() for line in file]
    return artist_names
```

```python
artists_filename = '../data/artists.txt'
artist_names = get_artist_names(artists_filename)
artist_names
```

```python
artist_filenames = [name.lower().replace(' ', '_') for name in artist_names]
artist_filenames
```

```python
artist_filepaths = ['../data/rt_data_dump_' + filename for filename in artist_filenames]
artist_filepaths
```
